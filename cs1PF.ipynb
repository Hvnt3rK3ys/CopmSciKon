<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;"> 0) üåü Introduction: The Hidden Geometry of Data </h2>

<p style="font-size: 1.6rem; line-height: 1.6;">
Topological Data Analysis (TDA) has emerged as a powerful discipline that allows us to uncover the intrinsic "shape" and hidden structures within complex datasets. Unlike traditional statistical methods that primarily focus on numerical metrics, TDA emphasizes the qualitative properties of data that remain invariant under continuous deformations, such as connectivity, holes, and voids. This unique perspective offers significant advantages in handling noise, high dimensionality, and heterogeneity‚Äîcommon characteristics of real-world data. [<a href="#cita18" target="_blank" style="color:#b5ff46; font-size:1.5rem">[18]</a>]</p>

<p style="font-size: 1.6rem; line-height: 1.6;">In this exposition, we will explore the historical and conceptual evolution of TDA, beginning with its fundamental applications across diverse fields. Subsequently, we will delve into its synergy with Machine Learning (ML), where TDA not only enhances existing models but also inspires new architectures. Finally, we will venture to the research frontier, examining how TDA is beginning to unveil the mysteries of Large Language Models (LLMs), offering tools for their interpretability and optimization. </p>

<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;"> 1) üìò Part A ‚Äì Fundamentals and Applications of TDA (Breadth of Use) </h2>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 1.1) What is Topological Data Analysis (TDA)? </h3>
<p style="font-size: 1.6rem; line-height: 1.6;">
Topological Data Analysis (TDA) is a field at the intersection of mathematics, statistics, and computer science that aims to extract meaningful information from the "shape" of data. Rooted in algebraic topology, TDA focuses on properties of data that are preserved under continuous transformations like stretching, bending, or twisting, but not tearing or gluing. These properties are known as topological invariants. [<a href="#cita18" target="_blank" style="color:#b5ff46; font-size:1.5rem">[18]</a>], [<a href="#cita7" target="_blank" style="color:#b5ff46; font-size:1.5rem">[7]</a>]

The core idea is that data points sampled from some underlying space often form complex geometric structures, and the global and local connectivity of these structures can reveal important insights. TDA provides a suite of tools to quantify these shape-based features. Two fundamental concepts underpin many TDA techniques:

*   **Simplicial Complexes:** These are mathematical structures used to represent the connectivity of data points at various scales. A $k$-simplex is the convex hull of $k+1$ affinely independent points. For example:
    *   A 0-simplex is a point: $[v_0]$
    *   A 1-simplex is an edge: $[v_0, v_1]$
    *   A 2-simplex is a triangle: $[v_0, v_1, v_2]$
    *   A 3-simplex is a tetrahedron: $[v_0, v_1, v_2, v_3]$

    An abstract simplicial complex $K$ on a finite set of vertices $V$ is a collection of non-empty subsets of $V$ (called simplices) such that if $\sigma \in K$ and $\tau \subseteq \sigma$, then $\tau \in K$. Furthermore, the intersection of any two simplices in $K$, if non-empty, must be a face of both. Simplicial complexes allow us to build a discrete representation of the underlying continuous space from which data is sampled. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>]

*   **Filtrations:** A filtration is a nested sequence of simplicial complexes, $K_0 \subseteq K_1 \subseteq K_2 \subseteq \dots \subseteq K_m = K$. This sequence is typically parameterized by a scale parameter $\epsilon$ (e.g., a distance threshold). As $\epsilon$ increases, more simplices are added to the complex, revealing structures at different resolutions. For example, in a Vietoris-Rips filtration $VR(P, \epsilon)$ built on a point cloud $P$, a simplex $[v_0, \dots, v_k]$ is included if all pairwise distances between its vertices are less than or equal to $\epsilon$:
    $$ VR(P, \epsilon) = \{ \sigma \subseteq P \mid \text{diam}(\sigma) \le \epsilon \} $$
    By varying $\epsilon$, we obtain a filtration. This process allows TDA to capture features that exist across multiple scales. [<a href="#cita15" target="_blank" style="color:#b5ff46; font-size:1.5rem">[15]</a>]

TDA complements traditional statistical and geometric approaches by providing qualitative, global insights into data structure, often revealing patterns missed by methods focused solely on local properties or numerical summaries.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 1.2) Why Use TDA? Key Advantages </h3>
<p style="font-size: 1.6rem; line-height: 1.6;">
TDA offers several intrinsic advantages that make it a valuable tool for data analysis:
<ul>
    <li><strong>Robustness to Noise and Perturbations:</strong> Topological features, especially those that persist across a wide range of scales (see Persistent Homology), are inherently stable. Small amounts of noise or minor perturbations in the data are less likely to significantly alter these global topological characteristics. [<a href="#cita1" target="_blank" style="color:#b5ff46; font-size:1.5rem">[1]</a>], [<a href="#cita16" target="_blank" style="color:#b5ff46; font-size:1.5rem">[16]</a>]</li>
    <li><strong>Capability to Handle High Dimensionality:</strong> TDA methods can be applied to data embedded in very high-dimensional spaces where direct visualization is impossible and many traditional geometric methods suffer from the "curse of dimensionality."</li>
    <li><strong>Discovery of Multiscale Structures:</strong> Through the concept of filtrations, TDA naturally explores data structures at multiple scales simultaneously, distinguishing transient noise from significant, persistent features.</li>
    <li><strong>Coordinate-Free and Metric-Agnostic (to some extent):</strong> The fundamental properties studied by topology are intrinsic to the "shape" of the data and do not depend on a specific coordinate system or embedding. While a metric is often needed to construct simplicial complexes (like Vietoris-Rips), TDA can be adapted to various notions of similarity or dissimilarity.</li>
</ul>
When compared to popular dimensionality reduction techniques like t-SNE or UMAP, TDA offers a different kind of insight. While t-SNE and UMAP are excellent for creating low-dimensional embeddings that preserve local neighborhood structures for visualization, they may distort global topological features. TDA, particularly Persistent Homology, aims to quantify these global features (like loops, voids) and their significance across scales, rather than just providing an embedding. [<a href="#cita1" target="_blank" style="color:#b5ff46; font-size:1.5rem">[1]</a>]
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 1.3) Central TDA Techniques </h3>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 1.3.1) Persistent Homology (PH) </h4>
<p style="font-size: 1.6rem; line-height: 1.6;">
Persistent Homology (PH) is arguably the most prominent tool in TDA. It tracks the "birth" and "death" of topological features (connected components, loops, voids, etc.) as a scale parameter $\epsilon$ (defining a filtration) varies. [<a href="#cita4" target="_blank" style="color:#b5ff46; font-size:1.5rem">[4]</a>], [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>]

The core idea is to study the homology groups $H_k(K_{\epsilon})$ of the simplicial complexes $K_{\epsilon}$ in a filtration. For a given dimension $k$:
<ul>
    <li>$H_0(K_{\epsilon})$ captures the connected components of $K_{\epsilon}$.
    <li>$H_1(K_{\epsilon})$ captures the loops (1-dimensional holes) in $K_{\epsilon}$.
    <li>$H_2(K_{\epsilon})$ captures the voids (2-dimensional holes) in $K_{\epsilon}$.
</ul>
The rank of $H_k(K_{\epsilon})$, denoted $\beta_k(\epsilon)$, is the $k$-th Betti number, which counts the number of $k$-dimensional holes at scale $\epsilon$.

A topological feature is said to be **born** at scale $\epsilon_{birth}$ if it appears in $K_{\epsilon_{birth}}$ but not in $K_{\epsilon}$ for $\epsilon < \epsilon_{birth}$. It **dies** at scale $\epsilon_{death}$ if it merges into an older, larger feature (for $k=0$) or gets "filled in" (for $k>0$) in $K_{\epsilon_{death}}$. The **persistence** of the feature is the length of the interval $[\epsilon_{birth}, \epsilon_{death})$, i.e., $p = \epsilon_{death} - \epsilon_{birth}$. Features with high persistence are considered significant, while those with low persistence are often attributed to noise.

The output of PH is typically visualized in two main ways:
<ul>
    <li><strong>Persistence Diagrams (PDs):</strong> A multiset of points $(b, d)$ in the 2D plane, where $b = \epsilon_{birth}$ and $d = \epsilon_{death}$ for each topological feature. Points far from the diagonal line $y=x$ correspond to highly persistent features.
        $$ PD_k = \{ (\epsilon_{birth}^{(i)}, \epsilon_{death}^{(i)}) \}_{i \in I_k} $$
        where $I_k$ is the index set of $k$-dimensional features.
    </li>
    <li><strong>Barcodes:</strong> A collection of horizontal intervals, where each interval $[b, d)$ represents the lifespan of a topological feature.
</ul>
PH is invaluable for quantifying the shape of data in a robust and multiscale manner.
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 1.3.2) Mapper Algorithm </h4>
<p style="font-size: 1.6rem; line-height: 1.6;">
The Mapper algorithm, introduced by Singh, M√©moli, and Carlsson [<a href="#cita4" target="_blank" style="color:#b5ff46; font-size:1.5rem">[4]</a>], is a technique for creating a simplified, graph-based representation of high-dimensional datasets. It provides a partial clustering of the data, visualized as a simplicial complex (often a graph). The steps are:
<ol>
    <li><strong>Filter Function (Lens):</strong> Choose one or more real-valued functions $f: X \to \mathbb{R}^k$ (often $k=1$ or $k=2$) defined on the data $X$. This function "filters" or projects the data into a lower-dimensional space. Examples include density estimators, PCA components, or domain-specific measures.</li>
    <li><strong>Covering the Range:</strong> The range (image) of $f$, $f(X)$, is covered by a collection of overlapping intervals (if $k=1$) or hyper-rectangles (if $k>1$). Let this cover be $\mathcal{U} = \{U_i\}_{i \in I}$.</li>
    <li><strong>Partial Clustering:</strong> For each set $U_i$ in the cover, consider its preimage $f^{-1}(U_i) = \{x \in X \mid f(x) \in U_i\}$. The data points in each $f^{-1}(U_i)$ are then clustered using a standard clustering algorithm (e.g., single-linkage clustering). Each resulting cluster $C_{ij}$ (the $j$-th cluster in the $i$-th preimage) becomes a node in the Mapper graph.</li>
    <li><strong>Graph Construction:</strong> An edge is drawn between two nodes (clusters) $C_{ij}$ and $C_{i'j'}$ if their corresponding sets of original data points have a non-empty intersection ($C_{ij} \cap C_{i'j'} \neq \emptyset$). This typically happens when the original cover elements $U_i$ and $U_{i'}$ overlap.
</ol>
The resulting Mapper graph provides a compressed representation of the data that highlights its connectivity and topological structure as viewed through the lens function $f$. It is particularly useful for visualization and exploratory data analysis of complex datasets. [<a href="#cita12" target="_blank" style="color:#b5ff46; font-size:1.5rem">[12]</a>]
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 1.3.3) Other Relevant Techniques (Brief Mention) </h4>
<p style="font-size: 1.6rem; line-height: 1.6;">
The work by Tu [<a href="#cita4" target="_blank" style="color:#b5ff46; font-size:1.5rem">[4]</a>] provides a comprehensive overview of efficient algorithms and applications within TDA, including detailed discussions on:
<ul>
    <li><strong>Contour Trees:</strong> These trees track the evolution of connected components of level sets of a scalar function defined on a simply connected domain (like an image). Nodes in the contour tree correspond to critical points of the function (minima, maxima, saddles), and edges represent regions of monotonic behavior. They are essentially Reeb graphs for simply connected domains.</li>
    <li><strong>Reeb Graphs:</strong> A generalization of contour trees, Reeb graphs capture the evolution of connected components of level sets of a scalar function defined on a manifold, which may have non-trivial topology (e.g., holes or tunnels). Nodes are critical points, and edges represent how these components merge or split as the function value changes.</li>
</ul>
These structures are fundamental in scientific visualization and shape analysis, providing skeletonized summaries of data.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 1.5) Challenges and Limitations of TDA in General Applications </h3>
<p style="font-size: 1.6rem; line-height: 1.6;">
Despite its strengths, TDA faces several challenges in general applications:
<ul>
    <li><strong>Computational Cost:</strong> Calculating persistent homology for large datasets, especially for higher-dimensional features or using dense constructions like Vietoris-Rips complexes, can be computationally intensive. The complexity can be, for instance, $O(N^3)$ or higher in the number of points $N$ for standard PH algorithms. [<a href="#cita4" target="_blank" style="color:#b5ff46; font-size:1.5rem">[4]</a>]</li>
    <li><strong>Interpretability of Topological Features:</strong> While TDA can reveal abstract topological features (e.g., a persistent 1-cycle), translating these mathematical constructs into meaningful, domain-specific insights often requires careful consideration and expertise. What does a specific "hole" signify in financial data versus biological data?</li>
    <li><strong>Parameter Selection:</strong> The results of TDA can be sensitive to the choice of various parameters, such as:
        <ul>
            <li>The metric used to define distances between data points.</li>
            <li>The type of filtration (e.g., Vietoris-Rips, Alpha complex, Cech complex).</li>
            <li>For Mapper: the choice of filter function(s), the cover of the range, and the clustering algorithm used on preimages.</li>
        </ul>
        Choosing appropriate parameters often requires domain knowledge and experimentation. [<a href="#cita1" target="_blank" style="color:#b5ff46; font-size:1.5rem">[1]</a>]
    </li>
</ul>
</p>

<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;"> 2) üî¨ Part B ‚Äì TDA in Synergy with Machine Learning (Enhancing Learning) </h2>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 2.1) Motivations for Integrating TDA and ML </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
The integration of Topological Data Analysis (TDA) and Machine Learning (ML) stems from the complementary strengths of both fields. While ML algorithms excel at learning patterns from data and making predictions, they often struggle with high-dimensional, noisy, or sparse data, and their decision-making processes can be opaque ("black boxes"). TDA, on the other hand, provides robust, multi-scale summaries of data shape that are invariant to certain transformations and can reveal global structures often missed by traditional ML feature extraction methods. [<a href="#cita7" target="_blank" style="color:#b5ff46; font-size:1.5rem">[7]</a>], [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>]

The key motivations for this synergy are:
<ul>
    <li><strong>Enhanced Feature Engineering:</strong> TDA can extract novel, robust features (topological signatures) that capture the intrinsic shape and connectivity of data. These features can augment traditional feature sets, providing richer information for ML models and potentially improving their performance.</li>
    <li><strong>Improved Generalization:</strong> By focusing on stable, global properties, TDA-derived features can help ML models generalize better to unseen data, especially when training data is limited or noisy.</li>
    <li><strong>Increased Interpretability:</strong> TDA can provide insights into the structure of the data itself and, when applied to analyze ML models (e.g., neural networks), can help to understand what these models have learned and how they make decisions. This is particularly relevant for complex models like deep neural networks. [<a href="#cita15" target="_blank" style="color:#b5ff46; font-size:1.5rem">[15]</a>]</li>
    <li><strong>Development of Novel Architectures:</strong> The principles of TDA are inspiring new ML architectures that are inherently "topology-aware," capable of processing and learning from topological information directly.</li>
</ul>
This combination aims to create more robust, interpretable, and powerful analytical tools.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 2.2) TDA as a Feature Engineering Tool for ML </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
One of the most direct ways to integrate TDA with ML is to use TDA outputs as features for standard ML classifiers or regressors. This requires transforming the primarily set-based or graphical outputs of TDA into fixed-length vector representations.
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 2.2.1) Vectorization of Persistence Diagrams and Barcodes </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
Persistence Diagrams (PDs) and barcodes, as discussed in Part A (Section 1.3.1), summarize the lifespan of topological features. Since ML algorithms typically require vector inputs, several methods have been developed to vectorize these TDA outputs:
<ul>
    <li>
        <strong>Persistence Images (PIs):</strong> Introduced by Adams et al. (2017) [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>, Sec 3.2.1], a PD is first transformed, often by mapping $(b, d)$ to $(b, d-b)$ (birth-persistence coordinates). Then, a Gaussian kernel is centered at each transformed point, and the sum of these kernels is evaluated on a grid. The resulting discretized surface is the persistence image, a fixed-size matrix (or vector) that can be fed into ML models. Mathematically, for a PD $D = \{ (b_i, d_i) \}$, the persistence surface $S(x, y)$ might be:
        $$ S(x,y) = \sum_{(b_i, d_i) \in D} w(b_i, d_i-b_i) \cdot K_{\sigma}((x,y) - (b_i, d_i-b_i)) $$
        where $K_{\sigma}$ is a 2D Gaussian kernel with standard deviation $\sigma$, and $w$ is an optional weighting function (e.g., proportional to persistence $d_i-b_i$). The PI is then obtained by integrating $S(x,y)$ over pixels of a grid.
    </li>
    <li>
        <strong>Persistence Landscapes (PLs):</strong> Introduced by Bubenik (2015) [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>, Sec 3.2.1], for each point $(b, d)$ in a PD, a "tent" function $\Lambda_{(b,d)}(t)$ is defined:
        $$ \Lambda_{(b,d)}(t) = \max(0, t-b) \text{ if } t \le \frac{b+d}{2} $$
        $$ \Lambda_{(b,d)}(t) = \max(0, d-t) \text{ if } t > \frac{b+d}{2} $$
        The $k$-th persistence landscape $\lambda_k(t)$ is then the $k$-th largest value of these tent functions at time $t$:
        $$ \lambda_k(t) = \text{kmax}_{ (b,d) \in D } \{ \Lambda_{(b,d)}(t) \} $$
        By sampling $\{\lambda_k(t)\}_{k=1}^K$ at a set of $t$ values, a vector representation is obtained.
    </li>
    <li>
        <strong>Betti Curves:</strong> As mentioned in Hensel et al. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>, Sec 3.2.1] and Zia et al. [<a href="#cita7" target="_blank" style="color:#b5ff46; font-size:1.5rem">[7]</a>, Sec 2.3], Betti curves $\beta_k(\epsilon)$ plot the $k$-th Betti number as a function of the filtration parameter $\epsilon$. Sampling this curve at discrete $\epsilon$ values yields a vector.
        $$ \vec{\beta_k} = [\beta_k(\epsilon_1), \beta_k(\epsilon_2), \dots, \beta_k(\epsilon_N)] $$
    </li>
    <li>
        <strong>Barcode Statistics:</strong> As used by Sagming et al. [<a href="#cita6" target="_blank" style="color:#b5ff46; font-size:1.5rem">[6]</a>], various statistical summaries of the birth times, death times, and persistences of features in a barcode can be computed (e.g., mean, median, standard deviation, min, max, sum of persistences). This directly yields a feature vector. For a set of persistence intervals $\{[b_i, d_i)\}_i$:
        $$ \text{Features} = [\text{mean}(b_i), \text{std}(d_i), \text{sum}(d_i-b_i), \dots] $$
    </li>
</ul>
These vectorizations allow the rich information from PDs to be used in standard ML pipelines. The choice of vectorization method can depend on the specific task and dataset.
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 2.2.2) Using Mapper Graphs as Input or for Feature Generation </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
The Mapper algorithm (Section 1.3.2) produces a graph (or more generally, a simplicial complex) that summarizes the dataset. This graph itself can be a source of features for ML:
<ul>
    <li><strong>Graph Kernels on Mapper Graphs:</strong> If the Mapper output is a graph, graph kernels can be used to compare different Mapper graphs, enabling their use in kernel-based ML methods like SVMs.</li>
    <li><strong>Graph-Theoretic Features:</strong> Standard graph-theoretic features can be extracted from the Mapper graph, such as the number of nodes, edges, connected components, degree distribution, centrality measures, or presence of specific motifs (e.g., flares or loops in the Mapper graph). These can then be used as a feature vector.</li>
    <li><strong>Node Features from Mapper:</strong> The nodes in a Mapper graph represent clusters of original data points. Features of these clusters (e.g., average value of original features within the cluster, size of the cluster) can be associated with the Mapper nodes and used in graph neural networks or other graph-based learning techniques.</li>
</ul>
The utility of Mapper here lies in its ability to provide a simplified yet informative representation of the data's global structure, which can then be quantified for ML.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 2.3) TDA for ML Model Improvement and Interpretation </h3>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 2.3.1) Topological Regularization </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
Topological regularization involves incorporating TDA-based penalties or constraints into the loss function of an ML model during training. The goal is to guide the learning process to favor models that exhibit desirable topological properties, often leading to improved generalization. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>, Sec 3.3.1], [<a href="#cita7" target="_blank" style="color:#b5ff46; font-size:1.5rem">[7]</a>, Sec 3.3]

For example, in a classification task, one might want the decision boundary to be topologically "simple" (e.g., having few connected components or no small, isolated regions). A TDA-based regularizer could penalize models that produce complex decision boundaries. Chen et al. (2019), cited in Hensel et al. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>], proposed such a regularizer based on the 0-dimensional persistence of the decision boundary.

Another approach involves ensuring that the topological structure of the model's internal representations (e.g., activations in a neural network layer) or its output space matches some known or desired topology. For instance, if data is known to lie on a manifold with a certain number of holes, a loss term could be added to encourage the learned latent space of an autoencoder to also have a similar number of holes. This is explored in the context of "Topological Autoencoders" by Moor et al. (2020), mentioned in Zia et al. [<a href="#cita7" target="_blank" style="color:#b5ff46; font-size:1.5rem">[7]</a>]. The loss might take the form:
$$ L_{total} = L_{task} + \lambda \cdot d_{topological}(PD_{latent}, PD_{target}) $$
where $L_{task}$ is the primary task loss (e.g., reconstruction error), $d_{topological}$ is a distance between persistence diagrams (e.g., Wasserstein or Bottleneck distance), and $PD_{target}$ is the target persistence diagram.
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 2.3.2) Interpreting "Black Box" ML Models with TDA </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
TDA can be used as a post-hoc analysis tool to probe and understand the internal workings of complex ML models, especially deep neural networks. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>, Sec 3.3.2], [<a href="#cita15" target="_blank" style="color:#b5ff46; font-size:1.5rem">[15]</a>]
<ul>
    <li><strong>Analyzing Activation Landscapes:</strong> The activations of neurons in a hidden layer of a neural network, in response to a set of inputs, form a point cloud in the activation space. TDA can be applied to this point cloud to study its topological structure. For instance, Naitzat et al. (2020), discussed in Ballester et al. [<a href="#cita15" target="_blank" style="color:#b5ff46; font-size:1.5rem">[15]</a>], studied how the topology of activations evolves layer by layer, observing a simplification (reduction in Betti numbers) as data propagates through the network, suggesting a disentangling of features.</li>
    <li><strong>Understanding Decision Boundaries:</strong> As mentioned in regularization, TDA can characterize the complexity of decision boundaries learned by classifiers. This can reveal if a model is overfitting (by creating overly complex boundaries) or underfitting.</li>
    <li><strong>Comparing Model Representations:</strong> TDA can be used to compare the internal representations learned by different models or by the same model at different training stages. This can help understand how representations evolve during training or how different architectures learn to represent data.</li>
</ul>
These analyses can provide valuable insights into what a model has learned, how it processes information, and why it makes certain predictions.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 2.4) Dedicated TDA-Based ML Models and Topological Classifiers </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
Beyond using TDA for feature engineering or post-hoc analysis, there is a growing interest in developing ML models that are intrinsically topological or designed specifically to operate on topological data structures.
<ul>
    <li>
        <strong>Topological Classifiers:</strong> Kindelan et al. [<a href="#cita14" target="_blank" style="color:#b5ff46; font-size:1.5rem">[14]</a>] proposed a classifier entirely based on TDA principles. Their method builds a filtered simplicial complex on the data and uses persistent homology to guide the selection of a "meaningful sub-complex" that approximates the data's topology. Label propagation is then performed on this sub-complex, leveraging topological neighborhoods (link and star of simplices) and filtration values as local weights. This approach aims to classify data, including imbalanced datasets, without relying on traditional ML stages after the TDA pipeline.
    </li>
    <li>
        <strong>Topology-Aware Neural Network Layers:</strong> Architectures like PersLay (Carri√®re et al., 2020, discussed in Hensel et al. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>]) and Persformer (Reinauer et al. [<a href="#cita3" target="_blank" style="color:#b5ff46; font-size:1.5rem">[3]</a>]) are neural network layers or architectures designed to take persistence diagrams (or other topological summaries) directly as input.
        <ul>
            <li><strong>PersLay:</strong> This layer learns a permutation-invariant representation of a persistence diagram by applying a learnable function to each point in the diagram and then aggregating these representations.
                $$ \text{PersLay}(D) = \text{op}(\{w(p) \cdot \phi(p) \mid p \in D\}) $$
                where $D$ is a PD, $w$ is a learnable weight function, $\phi$ is a learnable vector representation function, and $\text{op}$ is a permutation-invariant aggregation operator (e.g., sum, mean, max).
            </li>
            <li><strong>Persformer:</strong> This is a Transformer-based architecture that accepts persistence diagrams as input. It leverages the self-attention mechanism of Transformers to model interactions between points in a PD, making it suitable for learning complex relationships from topological data. It has shown strong performance on benchmark datasets.</li>
        </ul>
    </li>
</ul>
These models represent a deeper integration of topological concepts into the learning machinery itself.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 2.5) Performance and Comparisons: TDA+ML vs. Traditional ML </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
Several of your provided papers directly compare the performance of TDA-augmented ML models against traditional ML approaches:
<ul>
    <li><strong>Clustering:</strong> Combs & Bihl [<a href="#cita12" target="_blank" style="color:#b5ff46; font-size:1.5rem">[12]</a>] compared TDA (ToMATo and Mapper) with ten other clustering algorithms. On artificial 2D data, TDA (ToMATo) ranked third overall in accuracy. Mapper was highlighted for its unique visualization capabilities for high-dimensional data. This suggests TDA can be competitive, though sometimes more computationally expensive.</li>
    <li><strong>Graph Machine Learning:</strong> Taiwo et al. [<a href="#cita13" target="_blank" style="color:#b5ff46; font-size:1.5rem">[13]</a>] investigated TDA for graph classification. They found that while TDA did not significantly outperform graph kernels in terms of raw accuracy, it provided complementary information and enhanced robustness against outliers. Their proposed surrogate model integrating TDA aimed to improve interpretability.</li>
    <li><strong>COVID-19 Detection:</strong> Assaf et al. [<a href="#cita5" target="_blank" style="color:#b5ff46; font-size:1.5rem">[5]</a>] showed that using persistent homology features from lung CT scans as input to ML classifiers (Random Forest and SVM) achieved exceptional accuracy (97.5% for RF), outperforming models based on statistical/pixel values alone. This highlights the power of topological features in specific biomedical imaging tasks.</li>
    <li><strong>Customer Churn Prediction:</strong> Sagming et al. [<a href="#cita6" target="_blank" style="color:#b5ff46; font-size:1.5rem">[6]</a>] demonstrated that adding TDA barcode statistics as features to XGBoost significantly improved churn prediction accuracy (from 92.71% to 98.50%). This indicates that topological features can capture subtle patterns in customer behavior that are missed by traditional features.</li>
</ul>
These studies collectively show that while TDA+ML may not always provide a universal accuracy boost across all problems, it often offers advantages in terms of robustness, interpretability, or performance on specific types of data or tasks where shape and structure are crucial.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 2.6) Current Challenges in TDA-ML Integration </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
Despite the promising results, several challenges persist in the integration of TDA and ML:
<ul>
    <li><strong>Scalability of Topological Representations:</strong> Computing and vectorizing persistence diagrams for very large datasets or high-dimensional filtrations remains computationally demanding. While efficient algorithms exist for specific cases (e.g., low-dimensional PH, Ripser for Vietoris-Rips), general scalability is an ongoing research area.</li>
    <li><strong>Choice of Optimal TDA Representation for ML Tasks:</strong> There is no one-size-fits-all TDA representation (e.g., persistence images, landscapes, Betti curves, barcode statistics). The best choice often depends on the specific ML task, the nature of the data, and the ML algorithm being used. More systematic studies are needed to guide this selection. Hensel et al. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>] provide a survey of various methods.</li>
    <li><strong>Differentiability for End-to-End Training:</strong> A significant hurdle for deep integration is that many TDA constructions (like the standard persistence diagram computation) are not inherently differentiable with respect to the input data. This makes it difficult to train neural networks end-to-end with topological loss functions or layers that directly output persistence diagrams. Recent work (e.g., on differentiable topological layers like PersLay, or smoothed loss functions) is addressing this, but it remains an active area of research. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>, Sec 3.3.1], [<a href="#cita7" target="_blank" style="color:#b5ff46; font-size:1.5rem">[7]</a>, Sec 3.3]</li>
    <li><strong>High Computational Cost of Kernels:</strong> While topological kernels can be powerful, their computation can be expensive, often scaling quadratically or worse with the number of points in the persistence diagrams or the number of diagrams themselves, limiting their applicability to large datasets. [<a href="#cita8" target="_blank" style="color:#b5ff46; font-size:1.5rem">[8]</a>, Sec 3.2.2]</li>
</ul>
Addressing these challenges is key to unlocking the full potential of TDA in the ML landscape.
</p>

<h2 style="color:#8A26BD;font-size: 2rem; border-bottom: .5rem solid #71b12c;"> 3) üß† Part C ‚Äì TDA in the World of Large Language Models (The New Frontier) </h2>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 3.1) Why TDA for LLMs? Unveiling Latent Structures </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
Large Language Models (LLMs) have demonstrated remarkable capabilities in understanding and generating human language. However, their immense complexity, often involving billions of parameters and intricate architectures like Transformers, makes them "black boxes." Understanding their internal representations and decision-making processes is crucial for improving their reliability, interpretability, efficiency, and for mitigating biases. [<a href="#cita10" target="_blank" style="color:#b5ff46; font-size:1.5rem">[10]</a>]

TDA offers a novel lens to study the "geometry" and "shape" of the high-dimensional spaces learned by LLMs. The internal states of an LLM‚Äîsuch as token embeddings, hidden layer activations, or attention patterns‚Äîcan be viewed as point clouds in high-dimensional vector spaces. TDA can help answer questions like:
<ul>
    <li>How does the "shape" of these representations evolve as information flows through the layers of an LLM?</li>
    <li>Are there persistent topological structures (e.g., clusters, loops, voids) in these representation spaces that correspond to semantic or syntactic concepts?</li>
    <li>Can TDA help to identify redundancies or critical components within the LLM architecture?</li>
</ul>
By applying TDA, we aim to move beyond purely performance-based evaluations of LLMs and gain deeper insights into how they learn and represent language. This is a burgeoning field, with surveys like Uchendu & Le [<a href="#cita11" target="_blank" style="color:#b5ff46; font-size:1.5rem">[11]</a>] beginning to map out the landscape of TDA applications in Natural Language Processing (NLP).
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 3.2) Analyzing the Internal Representations of LLMs with TDA </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
TDA provides tools to probe various internal components of LLMs:
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 3.2.1) Topology of Embedding Spaces </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
LLMs represent words, phrases, or entire documents as dense vectors called embeddings. TDA can be used to study the geometric and topological structure of these embedding spaces.
<ul>
    <li><strong>Semantic Neighborhoods:</strong> Do words with similar meanings form topologically distinct clusters or manifolds in the embedding space? Uchendu & Le [<a href="#cita11" target="_blank" style="color:#b5ff46; font-size:1.5rem">[11]</a>] survey various approaches where TDA techniques like Persistent Homology or Mapper are applied to word or sentence embeddings (e.g., Word2Vec, GloVe, or Transformer-based embeddings) to identify semantic relationships, thematic clusters, or even the evolution of topics in a corpus. For example, one might construct a Vietoris-Rips complex on a set of sentence embeddings and compute its persistent homology. The 0-dimensional persistence diagram ($PD_0$) would reveal clustering structures (number of connected components at different scales), while $PD_1$ could indicate "loop-like" relationships between semantic concepts.</li>
    <li><strong>Language Phylogenies:</strong> Draganov and Skiena (2024), as discussed by Uchendu & Le [<a href="#cita11" target="_blank" style="color:#b5ff46; font-size:1.5rem">[11]</a>, Sec 3.3], used PH on word embeddings to investigate the "shape" of words and construct language phylogenetic trees, demonstrating that topological features can carry historical and structural linguistic information.</li>
    <li><strong>Word Manifolds:</strong> Fitz (2022), also in Uchendu & Le [<a href="#cita11" target="_blank" style="color:#b5ff46; font-size:1.5rem">[11]</a>, Sec 3.3], introduced the concept of a "word manifold," a simplicial complex built directly from strings of words, whose topology captures grammatical structure.</li>
</ul>
These studies aim to understand how LLMs organize semantic and syntactic information spatially.
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 3.2.2) Topological Analysis of Attention Mechanisms </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
Transformer-based LLMs heavily rely on self-attention mechanisms. Attention weights between tokens can be represented as a weighted directed graph for each attention head. TDA can be applied to these "attention graphs" to understand their structure and function. Perez & Reinauer [<a href="#cita9" target="_blank" style="color:#b5ff46; font-size:1.5rem">[9]</a>] in "The Topological BERT" pioneer this by:
<ol>
    <li>Transforming BERT's attention maps into attention graphs.</li>
    <li>Applying a filtration based on attention weights (e.g., $w_{ij}$ being the attention from token $i$ to token $j$).</li>
    <li>Computing persistent homology (e.g., 0-dimensional and 1-dimensional) for these filtered graphs.</li>
    <li>Using the resulting persistence diagrams (or vectorizations thereof) as input to a classifier for tasks like sentiment analysis or grammatical correctness.</li>
</ol>
This approach allows the model to learn from the topological patterns of attention, such as how information flows between tokens or which tokens form persistent clusters of attention. They found that topological features from attention graphs can perform comparably to or even outperform the standard BERT baseline on some tasks and exhibit higher robustness against adversarial attacks. Uchendu & Le [<a href="#cita11" target="_blank" style="color:#b5ff46; font-size:1.5rem">[11]</a>, Sec 4.2.3] also survey several works that extract TDA features from attention weights for tasks like deepfake text detection and grammatical acceptability judgment.
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 3.2.3) Investigating Hidden Layers and Activation Patterns </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
The activations of neurons in the hidden layers of an LLM form high-dimensional point clouds that change from layer to layer as input is processed.
<ul>
    <li><strong>Zigzag Persistence:</strong> Gardinazzi et al. [<a href="#cita10" target="_blank" style="color:#b5ff46; font-size:1.5rem">[10]</a>] introduce a novel framework using zigzag persistence to track the evolution of topological features (p-cycles) across the layers of an LLM. Unlike standard PH which studies a single filtration, zigzag persistence can handle a sequence of spaces and maps that may not be simple inclusions (e.g., $X_1 \leftrightarrow X_2 \leftrightarrow \dots \leftrightarrow X_L$).
        They define a "persistence similarity" metric based on the zigzag PDs to quantify how topological features transform or persist between layers. A zigzag filtration might be represented as:
        $$ \dots \xleftarrow{g_i} K_{i, i+1} \xrightarrow{f_{i+1}} K_{i+1} \xleftarrow{g_{i+1}} K_{i+1, i+2} \dots $$
        where $K_i$ is the simplicial complex built from activations at layer $i$, and $K_{i, i+1}$ is an intersection complex used to map features between layers. This allows tracking of features like connected components ($H_0$) or loops ($H_1$) as they evolve, merge, or disappear through the network's depth.
    </li>
    <li><strong>Persistence Scoring Function:</strong> Chauhan and Kaul (2022), cited by Uchendu & Le [<a href="#cita11" target="_blank" style="color:#b5ff46; font-size:1.5rem">[11]</a>, Sec 4.2.3], proposed a scoring metric based on the persistent homology of hidden BERT representations, which correlates with test accuracy and adversarial robustness.</li>
</ul>
These methods aim to characterize the information processing flow and functional specialization of different layers within LLMs.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 3.3) Applications of TDA in LLM Development and Understanding </h3>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 3.3.1) Model Pruning and Compression </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
LLMs are notoriously large. TDA can help identify redundancies. Gardinazzi et al. [<a href="#cita10" target="_blank" style="color:#b5ff46; font-size:1.5rem">[10]</a>] use their persistence similarity metric to identify and prune redundant layers in LLMs. If consecutive layers exhibit very high persistence similarity, it suggests that the transformation between them is not significantly altering the topological structure of the representations, and thus one of them might be redundant. Their experiments show comparable performance to state-of-the-art pruning methods on benchmark datasets. This is a significant application as it can lead to more efficient LLMs.
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 3.3.2) Interpretability and Explainability of LLM Decisions </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
Understanding why an LLM makes a particular prediction or generates specific text is a major challenge. TDA can contribute by:
<ul>
    <li><strong>Visualizing Representation Spaces:</strong> Mapper can be applied to the hidden states or attention patterns to create simplified graphical representations, potentially revealing how different inputs are processed or how concepts are clustered.</li>
    <li><strong>Identifying Salient Topological Features:</strong> As in Perez & Reinauer [<a href="#cita9" target="_blank" style="color:#b5ff46; font-size:1.5rem">[9]</a>], if certain topological features (e.g., specific loops or connected components in attention graphs) consistently appear for inputs leading to a particular output, these features might be indicative of the model's reasoning.</li>
</ul>
The goal is to link abstract topological structures to understandable aspects of the model's behavior.
</p>

<h4 style="color:#E67E22;font-size: 1.8rem;"> 3.3.3) Detection of Linguistic Phenomena and Biases </h4>
<p style="font-size: 1.2rem; line-height: 1.6;">
Uchendu & Le [<a href="#cita11" target="_blank" style="color:#b5ff46; font-size:1.5rem">[11]</a>, Sec 3 & 4.2.3] discuss how TDA can be applied to LLM embeddings or attention weights to probe for learned linguistic structures. For instance:
<ul>
    <li><strong>Syntactic Structures:</strong> Do LLMs implicitly learn parse trees or dependency relations? TDA might reveal hierarchical clustering in embeddings that mirrors syntactic structures. Port et al. (2018, 2022), mentioned in Uchendu & Le [<a href="#cita11" target="_blank" style="color:#b5ff46; font-size:1.5rem">[11]</a>, Sec 3.1.2], used TDA to analyze syntactic parameters across language families.</li>
    <li><strong>Semantic Relationships:</strong> TDA can explore how LLMs represent concepts like synonymy, antonymy, or analogy by examining the shape of clusters of related word embeddings.</li>
    <li><strong>Bias Detection:</strong> If embeddings of certain demographic groups form topologically distinct or inappropriately linked clusters, it might indicate learned biases.</li>
</ul>
This line of research uses TDA to "interrogate" LLMs about their linguistic knowledge.
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 3.4) Persformer and Other "TDA-Friendly" Architectures for Sequential Data </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
The Persformer architecture, introduced by Reinauer et al. [<a href="#cita3" target="_blank" style="color:#b5ff46; font-size:1.5rem">[3]</a>], is the first Transformer specifically designed to take persistence diagrams as input. While PDs are sets of points rather than sequences, the Transformer's self-attention mechanism is well-suited for modeling pairwise interactions between these (birth, death) points. Key aspects include:
<ul>
    <li><strong>Set Input:</strong> Transformers can be adapted to handle set-structured input by forgoing positional encodings, making them permutation-invariant, which is crucial for PDs.</li>
    <li><strong>Learning Vectorizations:</strong> Persformer learns a representation (vectorization) of the PD as part of its end-to-end training, rather than relying on a fixed, handcrafted vectorization method.</li>
    <li><strong>Universal Approximation:</strong> The paper proves that Persformer can uniformly approximate any continuous function on the space of persistence diagrams, highlighting its expressive power.</li>
</ul>
While Persformer was initially applied to synthetic and graph benchmark datasets, its principles could inspire or be adapted for processing sequences of topological features derived from text or other sequential data. <p style="font-size: 1.2rem; line-height: 1.6;">

<ul>

Further investigation into how topological properties affect LLM robustness and how TDA can be used to detect or defend against such attacks is a promising avenue.</li>

````python

%%time
from tqdm.auto import tqdm
import torch
import numpy as np
from torch.utils.data import Dataset
from gudhi.sklearn.cubical_persistence import CubicalPersistence
from gudhi.representations import DiagramSelector, PersistenceImage
from sklearn.pipeline import Pipeline
from sklearn.base import BaseEstimator, TransformerMixin

class CubicalTransformer(BaseEstimator, TransformerMixin):
    """Adaptador para garantizar compatibilidad con sklearn"""
    def __init__(self, homology_dimensions=[1]):
        self.homology_dimensions = homology_dimensions
        self.cubical = CubicalPersistence(
            homology_dimensions=homology_dimensions,
            n_jobs=-1
        )
    
    def fit(self, X, y=None):
        return self
    
    def transform(self, X):
        # Asegurar formato correcto para Gudhi
        if isinstance(X, np.ndarray) and X.ndim == 3:
            X = [x for x in X]  # Convertir a lista de arrays 2D
        return self.cubical.transform(X)

class TDAFeatureExtractor:
    def __init__(self, homology_dimensions=[1], resolution=(20, 20)):
        self.pipeline = Pipeline([
            ('cubical', CubicalTransformer(homology_dimensions)),
            ('selector', DiagramSelector(use=True, point_type="finite")),
            ('persistence_image', PersistenceImage(
                bandwidth=1.0,
                weight=lambda x: x[1]**2,
                resolution=resolution
            ))
        ])
        self.resolution = resolution
    
    def transform(self, images):
        # Conversi√≥n robusta de tensores a formato Gudhi
        if isinstance(images, torch.Tensor):
            images = images.cpu().numpy()
        
        # Manejo de diferentes formatos de entrada
        if images.ndim == 4:  # Batch [B, C, H, W]
            images = np.mean(images, axis=1)  # Convertir a escala de grises
        elif images.ndim == 3 and images.shape[0] == 3:  # Single imagen [C, H, W]
            images = np.mean(images, axis=0, keepdims=True)
        
        # Asegurar que los valores est√©n en rango [0,255]
        images = np.clip(images, 0, 255).astype(np.float32)
        
        return self.pipeline.transform(images)


class ImageTDADataset(Dataset):
    def __init__(self, dataset, tda_extractor):
        self.dataset = dataset
        self.extractor = tda_extractor
        self.tda_features = []
        
        for idx in tqdm(range(len(dataset)), desc="Precomputing TDA"):
            img, _ = dataset[idx]
            features = self.extractor.transform(img)
            self.tda_features.append(torch.FloatTensor(features))
    
    def __len__(self):
        return len(self.dataset)
    
    def __getitem__(self, idx):
        img, label = self.dataset[idx]
        return img, self.tda_features[idx], label

````

````python
%%time

import numpy as np
from tqdm.auto import tqdm
import torch
from torch.utils.data import Dataset
from gudhi.sklearn.cubical_persistence import CubicalPersistence
from gudhi.representations import DiagramSelector, PersistenceImage

class RobustTDAExtractor:
    def __init__(self, homology_dimensions=[1], resolution=(20, 20)):
        self.cubical = CubicalPersistence(
            homology_dimensions=homology_dimensions,
            input_type='top_dimensional_cells',
            n_jobs=-1
        )
        self.selector = DiagramSelector(use=True, point_type="finite")
        self.persistence_image = PersistenceImage(
            bandwidth=1.0,
            weight=lambda x: x[1]**2,
            resolution=resolution,
            im_range=[0, 256, 0, 256]  # A√±adido expl√≠citamente
        )
        self.resolution = resolution

    def transform(self, img_tensor):
        """M√©todo renombrado a transform para compatibilidad"""
        try:
            # Convertir tensor a numpy
            img_np = img_tensor[0].cpu().numpy() if isinstance(img_tensor, torch.Tensor) else img_tensor
            img_np = img_np.squeeze()

            # Normalizar a [0, 255]
            if img_np.max() <= 1.0:
                img_np = (img_np * 255).astype(np.float32)

            # Pipeline completo
            diagrams = self.cubical.transform([img_np])
            finite_diags = self.selector.transform(diagrams)
            features = self.persistence_image.transform(finite_diags)

            return features[0]
        except Exception:
            return np.zeros(np.prod(self.resolution))

    def process_image(self, img_tensor):
        """Convierte tensor de imagen a formato adecuado para Gudhi"""
        # Convertir a numpy y asegurar 2D
        img_np = img_tensor[0].cpu().numpy() if isinstance(img_tensor, torch.Tensor) else img_tensor
        img_np = img_np.squeeze()  # Eliminar dimensiones unitarias

        # Normalizar a [0, 255] y asegurar tipo float32
        if img_np.max() <= 1.0:
            img_np = (img_np * 255).astype(np.float32)
        else:
            img_np = img_np.astype(np.float32)

        return img_np

    def extract_features(self, img_np):
        """Procesamiento robusto paso a paso"""
        try:
            # Paso 1: Computar complejo c√∫bico
            diagrams = self.cubical.transform([img_np])

            # Manejar estructura de salida
            if isinstance(diagrams[0], list):
                combined_diagram = np.vstack([dim_diag for dim_diag in diagrams[0] if dim_diag.size > 0])
            else:
                combined_diagram = diagrams[0]

            # Paso 2: Filtrar puntos finitos
            finite_diagram = self.selector.transform([combined_diagram])[0]

            # Paso 3: Crear imagen de persistencia
            if finite_diagram.size == 0:
                return np.zeros(np.prod(self.resolution))

            return self.persistence_image.fit_transform([finite_diagram])[0]
        except Exception:
            return np.zeros(np.prod(self.resolution))

class TDADatasetFixed(Dataset):
    def __init__(self, dataset, tda_extractor):
        self.dataset = dataset
        self.extractor = tda_extractor
        self.tda_features = []

        for idx in tqdm(range(len(dataset)), desc="Precomputing TDA", leave=False):
            img, _ = dataset[idx]
            try:
                # Preprocesar imagen
                processed_img = self.extractor.process_image(img)

                # Extraer caracter√≠sticas
                features = self.extractor.extract_features(processed_img)

                self.tda_features.append(torch.FloatTensor(features))
            except Exception:
                dummy_feat = torch.zeros(np.prod(self.extractor.resolution))
                self.tda_features.append(dummy_feat)

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, idx):
        img, label = self.dataset[idx]
        return img, self.tda_features[idx], label

# Uso:
tda_extractor = RobustTDAExtractor(
    homology_dimensions=[1],  # Solo H1 para evitar problemas
    resolution=(20, 20)
)

# Inicializar extractor
tda_extractor = RobustTDAExtractor(homology_dimensions=[1], resolution=(20, 20))

# Crear datasets con TDA
train_dataset_tda = TDADatasetFixed(train_dataset, tda_extractor)
val_dataset_tda = TDADatasetFixed(val_dataset, tda_extractor)
test_dataset_tda = TDADatasetFixed(test_dataset, tda_extractor)

# DataLoaders con pin_memory para mejor rendimiento en GPU
train_loader = torch.utils.data.DataLoader(train_dataset_tda, batch_size=32, shuffle=True, pin_memory=True)
val_loader = torch.utils.data.DataLoader(val_dataset_tda, batch_size=32, shuffle=False, pin_memory=True)
test_loader = torch.utils.data.DataLoader(test_dataset_tda, batch_size=32, shuffle=False, pin_memory=True)
````

        <li><strong>Hybrid Model:</strong> Design a custom layer or branch in the CNN that processes topological features (more advanced, inspired by Persformer [<a href="#cita3" target="_blank" style="color:#b5ff46; font-size:1.5rem">[3]</a>] or Topological CNNs mentioned in Zia et al. [<a href="#cita7" target="_blank" style="color:#b5ff46; font-size:1.5rem">[7]</a>]).</li>
    </ul>
</p>

<h3 style="color:#71b12c;font-size: 2.2rem;"> 3.3) Loss Function, Optimizer, and Scheduler </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
    Specify the loss function (e.g., Categorical Crossentropy for multi-class classification).
    Choose an optimizer (e.g., Adam, SGD).
    Define any learning rate schedulers.
</p>

````python
%%time
# Define the optimizer
optimizer = optim.Adam(model.parameters(), lr=0.001)

# Define the loss function
criterion = nn.CrossEntropyLoss()

# Learning rate scheduler (optional)
scheduler = optim.lr_scheduler.StepLR(optimizer, step_size=10, gamma=0.1)
````

<h1 style="color:#0c0a0b;background-color:#71b12c;font-size: 4rem; text-align: center;padding:0.5rem;border-radius:5rem;"> 4) üèãÔ∏è Model Training ü¶æ </h1>
<h3 style="color:#71b12c;font-size: 2.2rem;"> 4.1) Training the Baseline Model(s) </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
    Compile and train the baseline CV model(s) using the training data generator.
    Use the validation set to monitor performance and for early stopping/hyperparameter tuning.
</p>

````python
%%time
"""TRADITIONAL CNN MODEL FOR CV CLUSTERING"""
# Training loop
num_epochs = 30
best_val_accuracy = 0

# Lists to store metrics for plotting
train_accuracies = []
val_accuracies = []
train_losses = []
val_losses = []

for epoch in range(num_epochs):
    model.train()  # Set model to training mode
    running_loss = 0.0
    correct = 0
    total = 0

    for images, labels in train_loader:
        images, labels = images.to(device), labels.to(device)

        optimizer.zero_grad()

        # Forward pass
        outputs = model(images)
        loss = criterion(outputs, labels)

        # Backward pass and optimize
        loss.backward()
        optimizer.step()

        # Update running loss and accuracy
        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

    # Calculate training accuracy and loss for this epoch
    train_accuracy = 100 * correct / total
    train_loss = running_loss / len(train_loader)

    train_accuracies.append(train_accuracy)  # Store training accuracy for this epoch
    train_losses.append(train_loss)  # Store training loss for this epoch
    print(f"Epoch {epoch+1}/{num_epochs}, Train Loss: {train_loss:.4f}, Train Accuracy: {train_accuracy:.2f}%")

    # Validation phase
    model.eval()  # Set model to evaluation mode
    correct = 0
    total = 0
    running_val_loss = 0.0
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = model(images)
            loss = criterion(outputs, labels)
            running_val_loss += loss.item()

            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    # Calculate validation accuracy and loss for this epoch
    val_accuracy = 100 * correct / total
    val_loss = running_val_loss / len(val_loader)

    val_accuracies.append(val_accuracy)  # Store validation accuracy for this epoch
    val_losses.append(val_loss)  # Store validation loss for this epoch
    print(f"Validation Accuracy: {val_accuracy:.2f}%, Validation Loss: {val_loss:.4f}")

    # Save the model with the best validation accuracy
    if val_accuracy > best_val_accuracy:
        best_val_accuracy = val_accuracy
        torch.save(model.state_dict(), 'best_model.pth')  # Save the model
        print("Model saved with better validation accuracy.")

    # Step the learning rate scheduler
    scheduler.step()
````

<h3 style="color:#71b12c;font-size: 2.2rem;"> 4.2) Training the TDA-Augmented Model(s) </h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
    If using TDA features as additional input, prepare the combined feature set.
    Train the TDA-augmented model(s), again using the validation set.
    If using late fusion, train the TDA-specific model separately.
</p>

````python

%%time
class CNNWithTDA(nn.Module):
    def __init__(self, base_cnn, tda_feature_size, num_classes):
        super().__init__()
        # Capas CNN
        self.cnn_features = nn.Sequential(
            *list(base_cnn.children())[:-1],
            nn.AdaptiveAvgPool2d((1, 1)),  # Coma a√±adida aqu√≠
        )

        # Capas para TDA con regularizaci√≥n
        self.tda_processor = nn.Sequential(
            nn.Linear(tda_feature_size, 128),
            nn.BatchNorm1d(128),
            nn.ReLU(),
            nn.Dropout(0.5),  # Coma a√±adida aqu√≠
        )

        # Clasificador final
        self.classifier = nn.Sequential(
            nn.Linear(2048 + 128, 512),  # ResNet50: 2048 features
            nn.BatchNorm1d(512),
            nn.ReLU(),
            nn.Dropout(0.5),
            nn.Linear(512, num_classes)
        )

    def forward(self, x_img, x_tda):
        # Extraer caracter√≠sticas CNN
        cnn_features = self.cnn_features(x_img).flatten(1)

        # Procesar caracter√≠sticas TDA
        tda_features = self.tda_processor(x_tda)

        # Combinar y clasificar
        combined = torch.cat((cnn_features, tda_features), dim=1)
        return self.classifier(combined)

# Inicializar modelo con pesos preentrenados actualizados
base_resnet = models.resnet50(weights=models.ResNet50_Weights.DEFAULT)
model = CNNWithTDA(
    base_cnn=base_resnet,
    tda_feature_size=400,  # 20x20 persistence image
    num_classes=3
).to(device)

# Congelar capas base
for param in model.cnn_features.parameters():
    param.requires_grad = False

# Optimizador con weight decay (L2 regularization)
optimizer = optim.AdamW([
    {'params': model.tda_processor.parameters()},
    {'params': model.classifier.parameters()}
], lr=0.001, weight_decay=1e-4)

# Loss function con label smoothing
criterion = nn.CrossEntropyLoss(label_smoothing=0.1)

# Scheduler con reducci√≥n en plateau
scheduler = optim.lr_scheduler.ReduceLROnPlateau(
    optimizer, mode='max', factor=0.1, patience=5, verbose=True)

def train_model(model, train_loader, val_loader, num_epochs=30):
    best_val_accuracy = 0
    history = {'train_loss': [], 'val_loss': [], 'train_acc': [], 'val_acc': []}

    for epoch in range(num_epochs):
        # Training
        model.train()
        running_loss = 0.0
        correct = 0
        total = 0

        for images, tda_vecs, labels in tqdm(train_loader, desc=f"Epoch {epoch+1}/{num_epochs}"):
            images, tda_vecs, labels = images.to(device), tda_vecs.to(device), labels.to(device)

            optimizer.zero_grad()
            outputs = model(images, tda_vecs)
            loss = criterion(outputs, labels)
            loss.backward()
            optimizer.step()

            running_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

        train_loss = running_loss / len(train_loader)
        train_accuracy = 100 * correct / total
        history['train_loss'].append(train_loss)
        history['train_acc'].append(train_accuracy)

        # Validation
        model.eval()
        val_loss = 0.0
        val_correct = 0
        val_total = 0
        all_preds = []
        all_labels = []

        with torch.no_grad():
            for images, tda_vecs, labels in val_loader:
                images, tda_vecs, labels = images.to(device), tda_vecs.to(device), labels.to(device)
                outputs = model(images, tda_vecs)
                loss = criterion(outputs, labels)
                val_loss += loss.item()
                _, predicted = torch.max(outputs.data, 1)
                val_total += labels.size(0)
                val_correct += (predicted == labels).sum().item()
                all_preds.extend(predicted.cpu().numpy())
                all_labels.extend(labels.cpu().numpy())

        val_loss /= len(val_loader)
        val_accuracy = 100 * val_correct / val_total
        history['val_loss'].append(val_loss)
        history['val_acc'].append(val_accuracy)

        # Calcular m√©tricas adicionales
        val_f1 = f1_score(all_labels, all_preds, average='weighted')
        val_precision = precision_score(all_labels, all_preds, average='weighted')
        val_recall = recall_score(all_labels, all_preds, average='weighted')

        print(f"\nEpoch {epoch+1}:")
        print(f"Train Loss: {train_loss:.4f} | Train Acc: {train_accuracy:.2f}%")
        print(f"Val Loss: {val_loss:.4f} | Val Acc: {val_accuracy:.2f}%")
        print(f"Val F1: {val_f1:.4f} | Val Precision: {val_precision:.4f} | Val Recall: {val_recall:.4f}")

        # Guardar mejor modelo
        if val_accuracy > best_val_accuracy:
            best_val_accuracy = val_accuracy
            torch.save({
                'epoch': epoch,
                'model_state_dict': model.state_dict(),
                'optimizer_state_dict': optimizer.state_dict(),
                'loss': val_loss,
                'accuracy': val_accuracy,
                'f1': val_f1,
                'precision': val_precision,
                'recall': val_recall
            }, 'best_model_with_tda.pth')
            print("¬°Mejor modelo guardado con m√©tricas de validaci√≥n!")

        # Actualizar scheduler
        scheduler.step(val_accuracy)

    return history

# Ejecutar entrenamiento
history = train_model(model, train_loader, val_loader, num_epochs=30)
````

<h3 style="color:#71b12c;font-size: 2.2rem;">5.5) TDA for Model Interpretability</h3>
<p style="font-size: 1.2rem; line-height: 1.6;">
    Topological Data Analysis (TDA) can be a powerful tool for interpreting the internal representations of trained Convolutional Neural Networks (CNNs). By applying TDA techniques such as the Mapper algorithm to the activation maps of specific layers, we can visualize and analyze the topological structures formed by these activations. This approach can help us understand how the model differentiates between various classes and whether distinct topological features emerge for different types of inputs.
</p>
<p style="font-size: 1.2rem; line-height: 1.6;">
    In our experiments, we observed that the CNN+TDA model had mixed performance compared to the classical CNN. While it showed improvements in classifying certain types of images, it also exhibited higher misclassification rates in other categories. For instance, the cross-error analysis revealed that the CNN+TDA model had more "Normal‚ÜíCOVID" and "Viral‚ÜíCOVID" misclassifications compared to the classical CNN. This suggests that while TDA can provide valuable insights and enhance model performance in some areas, it may also introduce complexities that affect overall accuracy.
</p>
<p style="font-size: 1.2rem; line-height: 1.6;">
    There are several potential reasons for the lower performance of the TDA-augmented model:
</p>
<ul style="font-size: 1.2rem; line-height: 1.6;">
    <li><strong>Complexity and Noise:</strong> TDA introduces additional complexity by incorporating topological features into the model. This complexity can sometimes lead to noise, making it harder for the model to generalize well across all classes.</li>
    <li><strong>Feature Representation:</strong> The way topological features are vectorized and integrated into the CNN architecture might not be optimal. Different methods of TDA vectorization and integration could potentially improve performance.</li>
    <li><strong>Training Dynamics:</strong> The training process for a CNN+TDA model might require different hyperparameters or optimization techniques compared to a classical CNN. The interplay between topological features and traditional CNN features could necessitate adjustments in learning rates, regularization, and other training parameters.</li>
    <li><strong>Data Characteristics:</strong> The effectiveness of TDA can depend on the specific characteristics of the dataset. For some types of data, topological features might not provide significant additional information beyond what is captured by traditional CNN features.</li>
</ul>
<p style="font-size: 1.2rem; line-height: 1.6;">
    Despite these challenges, TDA remains a promising approach for enhancing model interpretability and potentially improving performance in specific tasks. Future work could focus on refining TDA techniques, optimizing their integration with CNNs, and exploring their application to more complex vision tasks.
