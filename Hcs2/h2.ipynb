(0/20)</strong> Sea <span style="font-style: italic;">$\mathbf{a}$</span> un arreglo unidimensional de 1 o 0, donde la celda $i$ est谩 <em>viva</em> (o <em>muerta</em>) en el tiempo $t$ si $a_i(t) = 1$ (o $0$, respectivamente). $$

#### **1. Definiciones y Propiedades del Art铆culo**

Nos basamos en las definiciones presentadas por Stephen Wolfram en "Random Sequence Generation by Cellular Automata" (1986).

*   **Axioma 1: Espacio de Estados.** El estado de una celda $i$ en el tiempo $t$, denotado como $a_i(t)$, pertenece al conjunto $\{0, 1\}$.

*   **Definici贸n 1: Formulaci贸n L贸gica (Ecuaci贸n 3.1a).** La regla de actualizaci贸n se define como:
    $$ a'_i = a_{i-1} \text{ XOR } (a_i \text{ OR } a_{i+1}) $$
    (Para simplificar, usaremos $a'_i$ para $a_i(t+1)$ y $a_j$ para $a_j(t)$).

*   **Definici贸n 2: Formulaci贸n Aritm茅tica (Ecuaci贸n 3.1b).** El art铆culo postula la siguiente forma equivalente:
    $$ a'_i = (a_{i-1} + a_i + a_{i+1} + a_ia_{i+1}) \mod 2 $$

*   **Propiedad Fundamental (P谩gina 9):** El art铆culo especifica la correspondencia entre las operaciones l贸gicas y la aritm茅tica modular:
    *   **XOR** corresponde a la "adici贸n m贸dulo dos".
    *   **OR** corresponde a la "adici贸n Booleana".

#### **2. Lemas de Equivalencia Aritm茅tica**

A partir de la **Propiedad Fundamental**, establecemos los siguientes lemas para operar en el cuerpo de dos elementos, $\mathbb{F}_2$, donde la suma y la multiplicaci贸n son `mod 2`.

*   **Lema 1: Equivalencia de XOR.** Para $A, B \in \{0, 1\}$, la operaci贸n $A \text{ XOR } B$ es equivalente a la suma m贸dulo 2.
    $$ A \text{ XOR } B \equiv (A + B) \mod 2 $$
    *Justificaci贸n: Esto se deriva directamente de la definici贸n de XOR y de la tabla de sumar en $\mathbb{F}_2$.*

*   **Lema 2: Equivalencia de OR.** Para $A, B \in \{0, 1\}$, la operaci贸n $A \text{ OR } B$ es equivalente a la expresi贸n $(A + B + AB) \mod 2$.
    

| $A$ | $B$ | $A \text{ OR } B$ | $A+B$ | $AB$ | $(A+B+AB)\mod 2$ |
|:---:|:---:|:-----------------:|:-----:|:----:|:------------------------:|
| 0 | 0 | **0** | 0 | 0 | $0 \mod 2 = \textbf{0}$ |
| 0 | 1 | **1** | 1 | 0 | $1 \mod 2 = \textbf{1}$ |
| 1 | 0 | **1** | 1 | 0 | $1 \mod 2 = \textbf{1}$ |
| 1 | 1 | **1** | 2 | 1 | $3 \mod 2 = \textbf{1}$ |


    (El lema queda demostrado, ya que las columnas de resultados son id茅nticas).

#### **3. Demostraci贸n Formal de la Proposici贸n**

Comenzamos con la **Definici贸n 1 (Formulaci贸n L贸gica)** y aplicamos los lemas para llegar a la **Definici贸n 2 (Formulaci贸n Aritm茅tica)**.

1.  Sea la expresi贸n l贸gica de la Regla 30:
    $$ a'_i = a_{i-1} \text{ XOR } (a_i \text{ OR } a_{i+1}) $$

2.  Aplicamos el **Lema 2** al t茅rmino entre par茅ntesis $(a_i \text{ OR } a_{i+1})$ para convertirlo a su forma aritm茅tica:
    $$ a_i \text{ OR } a_{i+1} \equiv (a_i + a_{i+1} + a_ia_{i+1}) \mod 2 $$

3.  Sustituimos esta equivalencia en la expresi贸n original:
    $$ a'_i = a_{i-1} \text{ XOR } \left( (a_i + a_{i+1} + a_ia_{i+1}) \mod 2 \right) $$

4.  Ahora, aplicamos el **Lema 1** a la operaci贸n XOR principal. Sea $A = a_{i-1}$ y $B = (a_i + a_{i+1} + a_ia_{i+1})$. La expresi贸n se convierte en la suma `mod 2` de estos dos t茅rminos:
    $$ a'_i = \left( a_{i-1} + (a_i + a_{i+1} + a_ia_{i+1}) \right) \mod 2 $$

5.  Dado que la suma en `mod 2` es asociativa, podemos eliminar los par茅ntesis internos, obteniendo la expresi贸n final:
    $$ a'_i = (a_{i-1} + a_i + a_{i+1} + a_ia_{i+1}) \mod 2 $$

Esta expresi贸n final es id茅ntica a la **Definici贸n 2**, la formulaci贸n aritm茅tica presentada en el art铆culo.

        <strong>2. (0/20)</strong> Escribe un <em>script</em> en Python que evolucione la Regla 30 para una celda viva en $t=0$ hasta $t=1e6$ (un mill贸n de pasos). Requisitos:

```python
%%time
import numpy as np
import time
from tqdm import tqdm

class Rule30Automaton:
    """
    Implementa la evoluci贸n de un aut贸mata celular 1D seg煤n la Regla 30
    utilizando operaciones vectorizadas de NumPy para un rendimiento 贸ptimo.
    """

    def __init__(self, width: int, steps: int):
        """
        Inicializa el aut贸mata.

        Args:
            width (int): El ancho del universo celular (n煤mero de celdas).
                         Se recomienda un n煤mero impar para centrar el inicio.
            steps (int): El n煤mero total de pasos de tiempo a evolucionar.
        """
        if width % 2 == 0:
            print("Advertencia: Se recomienda un ancho impar para una mejor simetr铆a inicial.")
        
        self.width = width
        self.steps = steps
        # Usamos np.uint8 para eficiencia de memoria, ya que solo necesitamos 0s y 1s.
        self.grid = np.zeros(width, dtype=np.uint8)

    def initialize(self):
        """
        Prepara la condici贸n inicial del aut贸mata: una 煤nica celda viva en el centro.
        """
        # Limpiamos el grid por si se reutiliza la instancia
        self.grid[:] = 0
        # Establecemos la celda central en 1
        self.grid[self.width // 2] = 1
        print(f"Aut贸mata inicializado con un grid de {self.width} celdas.")
        print("Condici贸n inicial: Una 煤nica celda viva en el centro.")

    def _step(self):
        """
        Realiza un 煤nico paso de evoluci贸n. Esta es la funci贸n vectorizada.
        La l贸gica implementa: a_i' = a_{i-1} XOR (a_i OR a_{i+1})
        """
        # np.roll simula las condiciones de contorno peri贸dicas (el universo es un anillo)
        # a_{i-1} para todo el vector
        left_neighbors = np.roll(self.grid, 1)
        
        # a_{i+1} para todo el vector
        right_neighbors = np.roll(self.grid, -1)
        
        # Aplicamos la Regla 30 a todo el grid de una sola vez.
        # En NumPy, ^ es XOR y | es OR.
        self.grid = left_neighbors ^ (self.grid | right_neighbors)

    def evolve(self) -> np.ndarray:
        """
        Ejecuta la simulaci贸n completa para el n煤mero total de pasos.

        Returns:
            np.ndarray: El estado final del grid despu茅s de todos los pasos.
        """
        self.initialize()
        
        print(f"\nEvolucionando la Regla 30 por {self.steps} pasos...")
        # tqdm nos da una barra de progreso 煤til para procesos largos
        for _ in tqdm(range(self.steps), desc="Progreso de la Simulaci贸n"):
            self._step()
            
        print("Evoluci贸n completada.")
        return self.grid


if __name__ == "__main__":
    
    SIM_WIDTH = 2001 
    SIM_STEPS = 1_000_000  # Un mill贸n de pasos

    # 1. Crear la instancia del aut贸mata
    automaton = Rule30Automaton(width=SIM_WIDTH, steps=SIM_STEPS)

    # 2. Medir el tiempo y ejecutar la evoluci贸n
    start_time = time.time()
    final_state = automaton.evolve()
    end_time = time.time()

    # 3. Mostrar resultados
    duration = end_time - start_time
    
    # Contar las celdas vivas es un buen "checksum" para verificar el resultado
    live_cells = np.sum(final_state)

    print("\n--- Resultados de la Simulaci贸n ---")
    print(f"Tiempo total de ejecuci贸n: {duration:.4f} segundos")
    print(f"Ancho del grid: {SIM_WIDTH}")
    print(f"Pasos de tiempo: {SIM_STEPS}")
    print(f"N煤mero de celdas vivas en el estado final: {live_cells}")
    print(f"Estado final (primeras 70 celdas): \n{''.join(map(str, final_state[:70]))}...")
```

```python
%%time
import plotly.graph_objects as go
import numpy as np

class Rule30Visualizer(Rule30Automaton):
    def __init__(self, width: int, steps: int):
        super().__init__(width, steps)
        # Matriz de evoluci贸n que persiste en la instancia
        self.history = None

    def evolve_and_get_history(self) -> np.ndarray:
        """
        Evoluciona el aut贸mata y guarda todos los estados en una matriz.
        """
        self.initialize()
        self.history = np.zeros((self.steps, self.width), dtype=np.uint8)
        self.history[0] = self.grid.copy()
        
        for i in range(1, self.steps):
            self._step()
            self.history[i] = self.grid.copy()
        
        return self.history

    def plot(self, max_steps: int = 30):
        """
        Grafica la evoluci贸n del aut贸mata hasta un n煤mero de pasos dado.
        Args:
            max_steps (int): N煤mero de pasos a graficar (m谩x: self.steps).
        """
        if self.history is None:
            self.evolve_and_get_history()
        
        max_steps = min(max_steps, self.steps)
        
        fig = go.Figure(
            data=go.Heatmap(
                z=self.history[:max_steps],
                colorscale=[(0, "white"), (1, "black")],
                showscale=False
            )
        )
        fig.update_layout(
            title=f"Aut贸mata Celular Regla 30 (primeros {max_steps} pasos)",
            xaxis=dict(title="Celdas"),
            yaxis=dict(title="Tiempo (pasos)", autorange="reversed"),
            width=800,
            height=600  
        )
        fig.show()

if __name__ == "__main__":
    vis = Rule30Visualizer(width=61, steps=30)
    vis.plot(max_steps=30)
```

    <li>
        <strong>3. (0/10)</strong> Usa los valores de la columna central (donde estaba la celda viva en $t=0$), considerada ca贸tica, para:
        <ul>
            <li>Generar n煤meros aleatorios en el rango [0, 7].</li>
            <li>Graficar un histograma y analizar si la distribuci贸n es uniforme.</li>

```python
%%time
import numpy as np
import plotly.express as px

class Rule30RNG(Rule30Automaton):
    def __init__(self, width: int, steps: int):
        super().__init__(width, steps)
        self.center_col = width // 2
        self.history = None

    def evolve_and_get_history(self) -> np.ndarray:
        """
        Evoluciona el aut贸mata y guarda todos los estados en una matriz.
        """
        self.initialize()
        self.history = np.zeros((self.steps, self.width), dtype=np.uint8)
        self.history[0] = self.grid.copy()

        for i in range(1, self.steps):
            self._step()
            self.history[i] = self.grid.copy()

        return self.history

    def generate_numbers(self, n: int) -> np.ndarray:
        """
        Usa la columna central para generar n煤meros en [0,7].
        Junta bits de 3 en 3.
        """
        if self.history is None:
            self.evolve_and_get_history()
        
        column = self.history[:, self.center_col]

        max_numbers = len(column) // 3
        if n > max_numbers:
            print(f"锔 Aviso: solo se pueden generar {max_numbers} n煤meros con {len(column)} pasos. Usando {max_numbers}.")
            n = max_numbers

        trimmed = column[:n * 3]
        groups = trimmed.reshape(n, 3)

        # Convertir cada grupo de bits en entero [0,7]
        numbers = groups[:, 0] * 4 + groups[:, 1] * 2 + groups[:, 2] * 1
        return numbers



class Rule30RNGAnalyzer(Rule30RNG):
    def __init__(self, width: int, steps: int):
        super().__init__(width, steps)

    def plot_violin(self, n: int = 1000):
        """
        Genera n n煤meros pseudoaleatorios y muestra la distribuci贸n
        con un gr谩fico tipo violin.
        """
        numbers = self.generate_numbers(n)

        fig = px.violin(
            y=numbers,
            box=True,  # dibuja tambi茅n caja con mediana y cuartiles
            points="all",  # opcional: muestra todos los puntos (cuidado con 200k)
            title=f"Distribuci贸n de {len(numbers)} n煤meros generados con la Regla 30"
        )
        fig.update_layout(
            yaxis=dict(title="Valor (0-7)"),
            xaxis=dict(showticklabels=False),  # eje x es irrelevante aqu铆
            height=800,
            width=1200
        )
        fig.show()
        return numbers





if __name__ == "__main__":
    rng = Rule30RNGAnalyzer(width=2001, steps=500000)  
    rng.plot_violin(n=200000)


```
